### Step4: Data Analysis

```python
# Sorting the data set (ascending order)
df.sort_values(by="2022 Population").head(10)
```

```python
# Sorting the data set (descending order)
df.sort_values(by="2022 Population", ascending=False).head(10)
```

```python
#Setting the display format for floating-point numbers - here only two decimal places
#Lambda function lambda x: '%.2f' % x takes a floating-point number x as input and formats it using the string format specifier '%.2f', which represents a floating-point number with two decimal places
pd.set_option('display.float_format',lambda x:'%.2f'%x)
df
```

```python
# Further possibilities:
# Disable scientific notation:('display.float_format', None) -> will display the numbers in their full decimal form
```

```python
# To rearrange the order of the columns - first show all columns
df.columns
```

```python
# change order of columns - define new order and then reindex
new_order = ['Rank', 'CCA3', 'Country', 'Capital', 'Continent','Growth Rate','World Population Percentage','1970 Population','1980 Population','1990 Population','2000 Population','2010 Population','2015 Population','2020 Population','2022 Population','Area (km²)', 'Density (per km²)' ]
df = df.reindex(columns=new_order)
df
```

```python
# Calculate the total population by continent in 2022
population_by_continent = df.groupby('Continent')['2022 Population'].sum()

# Sort the population values in descending order
population_by_continent = population_by_continent.sort_values(ascending=False)

# Plot a bar chart of the population by continent
population_by_continent.plot(kind='bar')
plt.xlabel('Continent')
plt.ylabel('Population')
plt.title('Population by Continent')
plt.show()
```

```python
# refine the visualization
# Plot a bar chart of the population by continent & change the color to green
ax = population_by_continent.plot(kind='bar', color='green')

# Rotate the country names on the x-axis
ax.set_xticklabels(population_by_continent.index, rotation=45, ha='right')

# Remove the x-label 'Continent'
ax.set_xlabel('')

# Set the y-label and title
ax.set_ylabel('Population')
ax.set_title('Population by Continent')

plt.show()
```

### Step5: Explore Relationships between Variables

```python
# Scatter plot of population vs. area
df.plot.scatter(x='Area (km²)', y='2022 Population')
plt.xlabel('Area (km²)')
plt.ylabel('Population')
plt.title('Population vs. Area')
plt.show()
```

```python
# calculate the correlation
df.corr()
```

```python
# Check the correlation of numeric values only
# Calculate the correlation matrix of numeric columns only
numeric_corr = df.corr(numeric_only=True)

# Display the correlation matrix
print(numeric_corr)
```

```python
# Create a heatmap using seaborn - specifying the colormap (cmap) as 'coolwarm' and enabling annotations (annot=True) to show the correlation values on the heatmap
heatmap = sns.heatmap(numeric_corr, cmap='coolwarm', annot=True, annot_kws={"size": 6}) 
# Pass the annot_kws parameter with a dictionary specifying the desired size of the annotations

# Set the font size of the x-axis and y-axis labels
heatmap.set_yticklabels(heatmap.get_yticklabels(), size=6)
heatmap.set_xticklabels(heatmap.get_xticklabels(), size=6)

# Set the title of the heatmap
plt.title('Correlation Matrix')

# Display the heatmap
plt.show()
```

### Step6: Calculate Additional Statistics

```python
# Calculate the average population density in 2022
df['Density 2022 (per km²)'] = df['2022 Population'] / df['Area (km²)']
average_density = df['Density (per km²)'].mean()
df
```

```python
# Calculate the percentage of world population in 2022 for each country
df['World Population Percentage'] = (df['2022 Population'] / df['2022 Population'].sum()) * 100
df
```

```python
df.groupby('Continent').mean()
```

```python
#explicitly set numeric_only=True to indicate that you want to calculate the mean only for numeric columns
df.groupby('Continent').mean(numeric_only=True)
```

```python
# Groups the DataFrame df based on the unique values in the 'Continent' column. This creates separate groups for each unique continent
# calculates the mean of the numeric columns within each group - Non-numeric columns are excluded from the mean calculation
df2 = df.groupby('Continent').mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
df2.plot()
```

 Instead you can also use: df.groupby('Continent')[df.columns[5:13]].mean().sort_values(by="2022 Population", ascending=False)

```python
# Group df based on the unique values in the 'Continent' column
# Selects a subset of columns from the grouped DataFrame -  ['1980 Population', '1990 Population', '2000 Population', '2010 Population', '2015 Population', '2020 Population', '2022 Population'] for further calculations
# Calculates the mean of the selected numeric columns within each group. The numeric_only=True parameter ensures that only the numeric columns are considered in the calculation. Non-numeric columns are excluded from the mean calculation
# Sorts the resulting DataFrame based on the values in the "2022 Population" column in descending order. This rearranges the rows of the DataFrame based on the population in 2022, starting with the highest population
df2 = df.groupby('Continent')[['1980 Population',
       '1990 Population', '2000 Population', '2010 Population',
       '2015 Population', '2020 Population', '2022 Population']].mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
df3 = df2.transpose()
df3.plot(rot=45)
```
